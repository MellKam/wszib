```typescript
import { parse } from "jsr:@std/csv/parse";

const float32 = parse(await Deno.readTextFile("./sequence1_float32.csv"))
  .map(([index, value]) => [parseInt(index), parseFloat(value)]);
const float64 = parse(await Deno.readTextFile("./sequence1_float64.csv"))
  .map(([index, value]) => [parseInt(index), parseFloat(value)]);

const difference = float32.map((
  [index, value],
) => [index, value - float64[index][1]]);
console.log(difference);

declare global {
  const difference: [number, number][];
  const float32: [number, number][];
  const float64: [number, number][];
}
```

```typescript
import { document } from "jsr:@ry/jupyter-helper";
import * as Plot from "npm:@observablehq/plot";

Plot.plot({
  grid: true,
  color: {
    legend: true,
    domain: ["difference"],
    range: ["#efb118"],
  },
  marks: [
    Plot.ruleY([0]),
    Plot.lineY(
      difference.map(([index, value]) => ({
        index,
        value,
        name: "difference",
      })),
      { stroke: "name", x: "index", y: "value" },
    ),
  ],
  style: {
    background: "transparent",
  },
  document,
});
```
